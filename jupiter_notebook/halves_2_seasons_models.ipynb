```python
import matplotlib.pyplot as plt
import numpy as np
def print_test_vs_pred(y_test,prediction):
# Calculate the residuals (difference between actual and predicted)
  residuals = y_test.values.flatten() - prediction

  plt.figure(figsize=(8, 6))
  plt.scatter(y_test, prediction, alpha=0.5, )
  plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at 0
  plt.xlabel("Actual TotalxG")
  plt.ylabel("prediction")
  plt.title("Optimal Decision Tree Regressor")
  plt.grid(True)
  plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import r2_score
```

```python
grid_space = {'n_estimators' : range(2,5), 'learning_rate' : [0.1,0.2,0.5,0.6,0.7,0.8], 'max_depth' : range(2,6)}
```

```python
xgtree = xgb.XGBRegressor()
grid_search = GridSearchCV(xgtree, grid_space, cv=3, scoring='r2')
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
best_corr = grid_search.best_score_
pred = best_model.predict(X_test)
print(f"Best R-squared on training set (from Grid Search): {best_corr}")
print(f"R-squared on the test set with the best model: {r2_score(y_test, pred)}")
print_test_vs_pred(y_test, pred)
```

```python
# Plotting heatmaps
results = grid_search.cv_results_
r2_scores = results['mean_test_score']
learning_rates = grid_space['learning_rate']
n_estimators = grid_space['n_estimators']
max_depths = grid_space['max_depth']

for depth in max_depths:
    plt.figure(figsize=(10, 6))
    # Filter scores for the current max_depth
    depth_scores = r2_scores[results['param_max_depth'] == depth]
    # Reshape the scores into a grid corresponding to learning_rates and n_estimators
    heatmap_data = depth_scores.reshape(len(learning_rates), len(n_estimators))

    plt.imshow(heatmap_data, cmap='viridis', origin='lower',
               extent=[min(n_estimators) - 0.5, max(n_estimators) + 0.5,
                       min(learning_rates) - 0.5*(learning_rates[1]-learning_rates[0]),
                       max(learning_rates) + 0.5*(learning_rates[1]-learning_rates[0])],
               aspect='auto')
    plt.colorbar(label='Mean R-squared (CV)')
    plt.xlabel('n_estimators')
    plt.ylabel('learning_rate')
    plt.title(f'R-squared Heatmap for max_depth = {depth}')
    plt.xticks(n_estimators)
    plt.yticks(learning_rates)
    plt.show()
```

```python
y_flat_train = y_train.values.flatten()
y_flat_test = y_test.values.flatten()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Define the models and their parameter grids for Grid Search
models = {
    'Linear Regression': {
        'model': LinearRegression(),
        'params': {}
    },
    'Ridge': {
        'model': Ridge(),
        'params': {'alpha': [0.1, 1.0, 10.0, 100]}
    },
    'Lasso': {
        'model': Lasso(),
        'params': {'alpha': [0.1, 1.0, 10.0]}
    },
    'ElasticNet': {
        'model': ElasticNet(),
        'params': {'alpha': [0.1, 1.0, 10.0], 'l1_ratio': [0.1, 0.5, 0.9]}
    },
    'Decision Tree': {
        'model': DecisionTreeRegressor(),
        'params': {'max_depth': [None, 5, 10, 15]}
    },
    'Random Forest': {
        'model': RandomForestRegressor(),
        'params': {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]}
    },
    'SVR': {
        'model': SVR(),
        'params': {'C': [0.1, 1.0, 10.0], 'kernel': ['linear', 'rbf']}
    },
    'XGBoost': {
        'model': xgb.XGBRegressor(),
        'params': {'n_estimators': [2, 3, 4, 5], 'learning_rate': [0.5, 0.6, 0.7, 0.8], 'max_depth': [2, 3, 4, 5]}
    }
}

results = {}

for model_name, model_info in models.items():
    print(f"Running Grid Search for {model_name}...")
    grid_search = GridSearchCV(model_info['model'], model_info['params'], cv=4, scoring='r2')
    grid_search.fit(X_train, y_flat_train)
    best_model = grid_search.best_estimator_
    pred = best_model.predict(X_test)
    r2 = r2_score(y_test, pred)
    mse = mean_squared_error(y_test, pred)

    results[model_name] = {
        'best_params': grid_search.best_params_,
        'test_r2': r2,
        'test_mse': mse
    }
    print(f"Finished Grid Search for {model_name}. Test R-squared: {r2:.4f}, Test MSE: {mse:.4f}")
```

```python
import pandas as pd

results_df = pd.DataFrame(results).T
display(results_df)

best_model_name = results_df['test_r2'].idxmax()
print(f"\nBest performing model on the test set based on R-squared: {best_model_name}")
print("Results for the best model:")
display(results_df.loc[best_model_name])
```

```python
# Define the models and their parameter grids for Grid Search
models = {
    'Decision Tree': {
        'model': DecisionTreeRegressor(),
        'params': {'max_depth': [2,3,4,5]}}
}

results = {}

for model_name, model_info in models.items():
    print(f"Running Grid Search for {model_name}...")
    grid_search = GridSearchCV(model_info['model'], model_info['params'], cv=5, scoring='r2')
    grid_search.fit(X_train, y_flat_train)
    best_model = grid_search.best_estimator_
    pred = best_model.predict(X_test)
    r2 = r2_score(y_test, pred)
    mse = mean_squared_error(y_test, pred)

    results[model_name] = {
        'best_params': grid_search.best_params_,
        'test_r2': r2,
        'test_mse': mse
    }
    print(f"Finished Grid Search for {model_name}. Test R-squared: {r2:.4f}, Test MSE: {mse:.4f}")
```

```python
# Find the grid search object for Linear Regression
lr_grid_search = None
for model_name, model_info in models.items():
    if model_name == 'Linear Regression':
        lr_grid_search = GridSearchCV(model_info['model'], model_info['params'], cv=3, scoring='r2')
        # Fit the grid search again to get the best_estimator_
        lr_grid_search.fit(X_train, y_train)
        break

if lr_grid_search:
    best_lr_model = lr_grid_search.best_estimator_

    print("Linear Regression Intercept:", best_lr_model.intercept_)
    print("Linear Regression Coefficients:", best_lr_model.coef_)

    # Display coefficients with feature names
    feature_names = X_train.columns
    coefficients = best_lr_model.coef_[0] # Since it's a single target, coef_ is a 2D array

    for feature, coef in zip(feature_names, coefficients):
        print(f"{feature}: {coef}")
else:
    print("Linear Regression model not found in the defined models.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Assuming 'lr_grid_search' from the previous step holds the fitted GridSearchCV for Linear Regression
# and best_lr_model is the best estimator.
if lr_grid_search:
    lr_predictions = best_lr_model.predict(X_test)

    mae = mean_absolute_error(y_test, lr_predictions)
    mse = mean_squared_error(y_test, lr_predictions)
    rmse = np.sqrt(mse)

    print(f"Mean Absolute Error (MAE) for Linear Regression: {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE) for Linear Regression: {rmse:.4f}")
else:
    print("Linear Regression grid search object not found.")
```
