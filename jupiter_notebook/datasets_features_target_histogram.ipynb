```python
from google.colab import files
import pandas as pd
import io
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/full_games_dataset.csv")
```

```python
df.head()
```

```python
df = df.drop(['game_date','interval_duration'], axis=1)
```

```python
# df = df.drop(['opponent_team_xG','xG_diff', 'xPts'], axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop(['opponent_team_xG','xPts','xG_diff'], axis=1)
y = df['TotalxG']
```

```python
features_to_show = [feature for feature in X.columns if 'time' not in feature]
num_features = len(features_to_show)
fig, axes = plt.subplots(num_features, 2, figsize=(15, num_features * 3))

for i, feature in enumerate(features_to_show):
  axes[i, 0].hist(X[feature])
  axes[i, 0].set_ylabel('Count')
  axes[i, 0].set_xlabel(feature)
  axes[i, 0].set_title(f'Distribution of {feature}')

  axes[i, 1].scatter(X[feature], y)
  axes[i, 1].set_xlabel(feature)
  axes[i, 1].set_ylabel('TotalxG')
  axes[i, 1].set_title(f'{feature} vs. TotalxG')

plt.tight_layout()
plt.show()
```

```python
df_halves=pd.read_csv("/content/halves_dataset_23_25.csv")
```

```python
df_halves.head()
```

```python
df_halves = df_halves.drop(['interval_start','interval_duration','half'], axis=1)
```

```python
df_halves = df_halves[df_halves['zone_6_distance_attackers']<8000]
```

```python
df_halves = df_halves[df_halves['decelerations']<700]
```

```python
df_halves.shape
```

```python
X_halves = df_halves.drop(['TotalxG'], axis=1)
y_halves = df_halves['TotalxG']
```

```python
features_to_show = [feature for feature in X_halves.columns if 'time' not in feature]
num_features = len(features_to_show)
fig, axes = plt.subplots(num_features, 2, figsize=(15, num_features * 3))

for i, feature in enumerate(features_to_show):
    axes[i, 0].hist(X_halves[feature])
    axes[i, 0].set_ylabel('Count')
    axes[i, 0].set_xlabel(feature)
    axes[i, 0].set_title(f'Distribution of {feature}')

    axes[i, 1].scatter(X_halves[feature], y_halves)
    axes[i, 1].set_xlabel(feature)
    axes[i, 1].set_ylabel('TotalxG')
    axes[i, 1].set_title(f'{feature} vs. TotalxG')

plt.tight_layout()
plt.show()
```
